```python
import pandas as pd
import json
#pd.set_option('display.max_rows', 100)
#pd.set_option('display.max_columns', None)
```

```python
classfile_in = 'hubbles-hot-stars-classifications.csv'
classifications = pd.read_csv(classfile_in)
len(classifications)
```

```python
# Expanding JSON fields into dictionaries
classifications['metadata_json'] = [json.loads(q) for q in classifications.metadata]
classifications['annotations_json'] =[json.loads(q) for q in classifications.annotations]
classifications['subject_data_json'] = [json.loads(q) for q in classifications.subject_data]
```

```python
# I think this will be a helpful column in the future when extracting additional info from classifications
classifications['n_annotations'] = [ len(q) for q in classifications.annotations_json ]

# Here is an edited version where I extract Y/N response for first answer for full classifications structure.
# Probably better to make these new columns for the full dataset before slicing by workflow, etc.
classifications['hadetect_answer'] =([q[0]['value'] for q in classifications.annotations_json])
```

```python
# You already have the subject_id info you need for aggregation in the subject_ids column of classifications structure.
classifications['subject_ids']
```

```python
# I single out one random classification to demonstrated how to extract data within subject_data_json.
test=classifications[classifications['classification_id'] == 25611067]
```
